```
data=pd.read_csv("core_dataset.csv")
data.head()
```

```
data.drop(['Employee Number','Performance Score','State','Zip','DOB','MaritalDesc','CitizenDesc','Hispanic/Latino','RaceDesc','Reason For Term','Department','Position','Pay Rate','Manager Name'],axis=1, inplace=True)
data.head()
```

```
data.dropna(axis=0,how='any',subset=['Date of Termination'],inplace=True)
data.head()
```

```
data['Hire Year'] = pd.DatetimeIndex(data['Date of Hire']).year
data['Hire Year']
```

```
data.drop('Date of Hire',axis=1,inplace=True)
data.head()
```

```
data.rename(columns={"Hire Year": "HireYear"},inplace=True)
```
